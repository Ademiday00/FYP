```python
# Install packages
!pip install transformers tensorflow scikit-learn pandas seaborn tf-keras

# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, auc
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModel, create_optimizer

# 1. Load dataset
csv_path = "/content/drive/MyDrive/clinical_info.csv"
data = pd.read_csv(csv_path)

# 2. Drop irrelevant columns
drop_cols = ['ID', 'SIDE', 'FILENAME']
data = data.drop(columns=drop_cols)

# 3. Encode categorical columns
categorical_cols = ['FREQUENT PAIN', 'SURGERY', 'RISK', 'SXKOA', 'SWELLING', 'BENDING FULLY', 'SYMPTOMATIC', 'CREPITUS']
for col in categorical_cols:
    data[col] = LabelEncoder().fit_transform(data[col].astype(str))

# 4. Create text input for model
def create_text(row):
    return (
        f"Age {row['AGE']} years old, BMI {row['BMI']}, "
        f"frequent pain {row['FREQUENT PAIN']}, surgery {row['SURGERY']}, "
        f"risk factor {row['RISK']}, swelling {row['SWELLING']}, "
        f"symptomatic {row['SYMPTOMATIC']}, crepitus {row['CREPITUS']}."
    )

data['TEXT'] = data.apply(create_text, axis=1)

# 5. Convert KOOS PAIN SCORE into classes
def pain_score_to_class(score):
    if score <= 33:
        return 0  # Severe
    elif score <= 66:
        return 1  # Moderate
    else:
        return 2  # Mild

data['PAIN_CLASS'] = data['KOOS PAIN SCORE'].apply(pain_score_to_class)

# 6. Prepare features and labels
texts = data['TEXT'].tolist()
labels = data['PAIN_CLASS'].values

# 6.5 Reduce dataset size (optional)
sample_size = 200  # Change to your desired number
texts = texts[:sample_size]
labels = labels[:sample_size]

# 7. Train-test split
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# 8. Load BioBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-base-cased-v1.1")
biobert = TFAutoModel.from_pretrained("dmis-lab/biobert-base-cased-v1.1", from_pt=True)

# 9. Tokenize text
def tokenize_texts(texts, tokenizer, max_len=128):
    return tokenizer(
        texts,
        padding='max_length',
        truncation=True,
        max_length=max_len,
        return_tensors="tf"
    )

train_encodings = tokenize_texts(X_train, tokenizer)
test_encodings = tokenize_texts(X_test, tokenizer)

# 10. Build Hybrid Model
input_ids = tf.keras.Input(shape=(128,), dtype=tf.int32, name="input_ids")
attention_mask = tf.keras.Input(shape=(128,), dtype=tf.int32, name="attention_mask")
embedding = biobert(input_ids, attention_mask=attention_mask)[0]

lstm = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True))(embedding)
cnn = tf.keras.layers.Conv1D(64, kernel_size=3, activation='relu', padding='same')(lstm)
pooling = tf.keras.layers.GlobalMaxPooling1D()(cnn)
dense = tf.keras.layers.Dense(32, activation='relu')(pooling)
dropout = tf.keras.layers.Dropout(0.3)(dense)
output = tf.keras.layers.Dense(3, activation='softmax')(dropout)

model = tf.keras.Model(inputs=[input_ids, attention_mask], outputs=output)

steps_per_epoch = len(X_train) // 8
num_train_steps = steps_per_epoch * 10
optimizer, _ = create_optimizer(init_lr=1e-4, num_train_steps=num_train_steps, num_warmup_steps=100)

model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Early stopping
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)

# 11. Train model
history = model.fit(
    x={
        "input_ids": train_encodings["input_ids"],
        "attention_mask": train_encodings["attention_mask"],
    },
    y=y_train,
    validation_data=(
        {
            "input_ids": test_encodings["input_ids"],
            "attention_mask": test_encodings["attention_mask"],
        },
        y_test
    ),
    epochs=20,
    batch_size=8,
    callbacks=[early_stop]
)

# 12. Evaluate model
loss, accuracy = model.evaluate(
    {
        "input_ids": test_encodings["input_ids"],
        "attention_mask": test_encodings["attention_mask"],
    },
    y_test
)
print(f"\nTest Loss: {loss:.4f}")
print(f"\nTest Accuracy: {accuracy:.4f}")


# 13. Predictions
y_pred_probs = model.predict({
    "input_ids": test_encodings["input_ids"],
    "attention_mask": test_encodings["attention_mask"],
})
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# 14. Confusion Matrix and Report
cm = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Severe', 'Moderate', 'Mild'], yticklabels=['Severe', 'Moderate', 'Mild'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_classes, target_names=['Severe', 'Moderate', 'Mild']))

# 15. Plot Loss and Accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')
plt.legend()

plt.tight_layout()
plt.show()

# 16. AUC-ROC Curve (Multi-class)
lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)
y_pred_probs_bin = y_pred_probs

fpr = {}
tpr = {}
roc_auc = {}
class_names = ['Severe', 'Moderate', 'Mild']

for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_probs_bin[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']

for i, color in zip(range(3), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve for {class_names[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC AUC Curve')
plt.legend(loc="lower right")
plt.grid()
plt.show()

macro_auc = roc_auc_score(y_test_bin, y_pred_probs_bin, average='macro')
weighted_auc = roc_auc_score(y_test_bin, y_pred_probs_bin, average='weighted')
print(f"Macro AUC: {macro_auc:.4f}")
print(f"Weighted AUC: {weighted_auc:.4f}")
```

```python
# Install packages
!pip install transformers tensorflow scikit-learn pandas seaborn tf-keras

# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, auc
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModel, create_optimizer

# 1. Load dataset
csv_path = "/content/drive/MyDrive/clinical_info.csv"
data = pd.read_csv(csv_path)

# 2. Drop irrelevant columns
drop_cols = ['ID', 'SIDE', 'FILENAME']
data = data.drop(columns=drop_cols)

# 3. Encode categorical columns
categorical_cols = ['FREQUENT PAIN', 'SURGERY', 'RISK', 'SXKOA', 'SWELLING', 'BENDING FULLY', 'SYMPTOMATIC', 'CREPITUS']
for col in categorical_cols:
    data[col] = LabelEncoder().fit_transform(data[col].astype(str))

# 4. Create text input for model
def create_text(row):
    return (
        f"Age {row['AGE']} years old, BMI {row['BMI']}, "
        f"frequent pain {row['FREQUENT PAIN']}, surgery {row['SURGERY']}, "
        f"risk factor {row['RISK']}, swelling {row['SWELLING']}, "
        f"symptomatic {row['SYMPTOMATIC']}, crepitus {row['CREPITUS']}."
    )

data['TEXT'] = data.apply(create_text, axis=1)

# 5. Convert KOOS PAIN SCORE into classes
def pain_score_to_class(score):
    if score <= 33:
        return 0  # Severe
    elif score <= 66:
        return 1  # Moderate
    else:
        return 2  # Mild

data['PAIN_CLASS'] = data['KOOS PAIN SCORE'].apply(pain_score_to_class)

# 6. Prepare features and labels
texts = data['TEXT'].tolist()
labels = data['PAIN_CLASS'].values

# 6.5 Reduce dataset size (optional)
sample_size = 200  # Change to your desired number
texts = texts[:sample_size]
labels = labels[:sample_size]

# 7. Train-test split
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.25, random_state=42)

# 8. Load BioBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-base-cased-v1.1")
biobert = TFAutoModel.from_pretrained("dmis-lab/biobert-base-cased-v1.1", from_pt=True)

# 9. Tokenize text
def tokenize_texts(texts, tokenizer, max_len=64):
    return tokenizer(
        texts,
        padding='max_length',
        truncation=True,
        max_length=max_len,
        return_tensors="tf"
    )

train_encodings = tokenize_texts(X_train, tokenizer)
test_encodings = tokenize_texts(X_test, tokenizer)

# 10. Build Hybrid Model
input_ids = tf.keras.Input(shape=(64,), dtype=tf.int32, name="input_ids")
attention_mask = tf.keras.Input(shape=(64,), dtype=tf.int32, name="attention_mask")
embedding = biobert(input_ids, attention_mask=attention_mask)[0]

lstm = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True))(embedding)
cnn = tf.keras.layers.Conv1D(64, kernel_size=3, activation='relu', padding='same')(lstm)
pooling = tf.keras.layers.GlobalMaxPooling1D()(cnn)
dense = tf.keras.layers.Dense(32, activation='relu')(pooling)
dropout = tf.keras.layers.Dropout(0.3)(dense)
output = tf.keras.layers.Dense(3, activation='softmax')(dropout)

model = tf.keras.Model(inputs=[input_ids, attention_mask], outputs=output)

steps_per_epoch = len(X_train) // 8
num_train_steps = steps_per_epoch * 10
optimizer, _ = create_optimizer(init_lr=1e-4, num_train_steps=num_train_steps, num_warmup_steps=100)

model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Early stopping
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)

# 11. Train model
history = model.fit(
    x={
        "input_ids": train_encodings["input_ids"],
        "attention_mask": train_encodings["attention_mask"],
    },
    y=y_train,
    validation_data=(
        {
            "input_ids": test_encodings["input_ids"],
            "attention_mask": test_encodings["attention_mask"],
        },
        y_test
    ),
    epochs=20,
    batch_size=8,
    callbacks=[early_stop]
)

# 12. Evaluate model
loss, accuracy = model.evaluate(
    {
        "input_ids": test_encodings["input_ids"],
        "attention_mask": test_encodings["attention_mask"],
    },
    y_test
)
print(f"\nTest Loss: {loss:.4f}")
print(f"\nTest Accuracy: {accuracy:.4f}")


# 13. Predictions
y_pred_probs = model.predict({
    "input_ids": test_encodings["input_ids"],
    "attention_mask": test_encodings["attention_mask"],
})
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# 14. Confusion Matrix and Report
cm = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Severe', 'Moderate', 'Mild'], yticklabels=['Severe', 'Moderate', 'Mild'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_classes, target_names=['Severe', 'Moderate', 'Mild']))

# 15. Plot Loss and Accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')
plt.legend()

plt.tight_layout()
plt.show()

# 16. AUC-ROC Curve (Multi-class)
lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)
y_pred_probs_bin = y_pred_probs

fpr = {}
tpr = {}
roc_auc = {}
class_names = ['Severe', 'Moderate', 'Mild']

for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_probs_bin[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']

for i, color in zip(range(3), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve for {class_names[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC AUC Curve')
plt.legend(loc="lower right")
plt.grid()
plt.show()

macro_auc = roc_auc_score(y_test_bin, y_pred_probs_bin, average='macro')
weighted_auc = roc_auc_score(y_test_bin, y_pred_probs_bin, average='weighted')
print(f"Macro AUC: {macro_auc:.4f}")
print(f"Weighted AUC: {weighted_auc:.4f}")
```
